```
import streamlit as st
import google.generativeai as genai

# 1. Konfigurasi Halaman Aplikasi
# Mengubah judul di tab peramban dan ikon
st.set_page_config(page_title="Tutor AI Bimbel TPA", page_icon="📝")

# Mengubah judul besar di halaman utama web
st.title("TUTOR AI BIMBEL TPA")

# 2. Pengaturan Keamanan API Key
api_key = st.secrets["GEMINI_API_KEY"]
genai.configure(api_key=api_key)

# 3. Penyusunan Instruksi Sistem (Diperbarui untuk TPA)
instruksi_sistem = """
Anda adalah asisten bimbingan belajar khusus untuk Tes Potensi Akademik (TPA).
Tugas Anda adalah memberikan soal latihan terstruktur mencakup logika, verbal, dan numerik.
Anda harus mengevaluasi jawaban pengguna dan memberikan pembahasan akademis yang akurat
untuk mengukur dan mempersiapkan kesiapan ujian pengguna secara objektif.
"""

# 4. Inisialisasi Model Bahasa
model = genai.GenerativeModel(
    model_name="gemini-1.5-flash",
    system_instruction=instruksi_sistem
)

# 5. Manajemen Memori Percakapan (Session State)
if "chat_session" not in st.session_state:
    st.session_state.chat_session = model.start_chat(history=[])

for message in st.session_state.chat_session.history:
    role = "user" if message.role == "user" else "assistant"
    with st.chat_message(role):
        st.markdown(message.parts[0].text)

# 6. Kolom Masukan Pengguna (Teks petunjuk diperbarui)
user_input = st.chat_input("Ketik pertanyaan atau minta soal latihan TPA di sini...")

if user_input:
    with st.chat_message("user"):
        st.markdown(user_input)

    with st.chat_message("assistant"):
        response = st.session_state.chat_session.send_message(user_input)
        st.markdown(response.text)
```
